```python
# read testcsv and remove user_id column and save it using pandas 
import pandas as pd
df = pd.read_csv('test.csv')
df.drop('user_id', axis=1, inplace=True)
df.to_csv('test.csv', index=False)
```

```python
# read from the same csv and read image (base64) column and decode it and save it as image with name as `cid`
import base64
import os
os.system("rm -r data")
os.mkdir('data')
df = pd.read_csv('test.csv')
for index, row in df.iterrows():
    imgdata = base64.b64decode(row['image'].split(",")[1])
    format = row['image'].split(";")[0].split("/")[1]
    filename = f"data/{row['cid']}_{row['id']}.{format}"
    with open(filename, "wb") as f:
        f.write(imgdata)
        print('image saved as', filename)
```

```python
# print all cids with no repeat
print("\n".join(df['cid'].unique().astype(str)))
```

```python
block_list = [962, 222, 313, 280]
# moeve all images with cid in block_list from data to block folder
try:
    os.mkdir('block')
except FileExistsError:
    os.system("mv block/* data")
for index, row in df.iterrows():
    if row['cid'] in block_list:
        os.system(f"mv data/{row['cid']}_{row['id']}.{format} block")
```
